# Phase 2: Data Models & Exceptions Test
This notebook verifies that all Pydantic models in `src.core.models` can be instantiated and that validation works as expected.

```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.core.models import (
    Source,
    Chunk,
    Query,
    RetrievedChunk,
    Citation,
    LLMResponse,
    ChatMessage,
    Session,
)
from src.core.exceptions import MiniNotebookLMError, IngestionError
from datetime import datetime
import pydantic

print("Imports successful!")
```

## 1. Test Source Model

```python
try:
    source = Source(
        id="source_1",
        title="Test Document",
        source_type="pdf",
        file_path="/path/to/test.pdf",
    )
    print(f"Source instantiation successful: {source.id}")

    # Test validation error
    try:
        invalid_source = Source(id="s2", title="Error", source_type="invalid_type")
    except pydantic.ValidationError as e:
        print("Source validation works! Caught invalid source_type.")
except Exception as e:
    print(f"Source test failed: {e}")
```

## 2. Test Chunk Model

```python
try:
    chunk = Chunk(
        id="chunk_1",
        source_id="source_1",
        content="This is a test chunk.",
        modality="text",
        embedding=[0.1, 0.2, 0.3],
    )
    print(f"Chunk instantiation successful: {chunk.id}")
except Exception as e:
    print(f"Chunk test failed: {e}")
```

## 3. Test LLMResponse & Citation

```python
try:
    citation = Citation(
        text="Sample text", source_id="source_1", chunk_id="chunk_1", confidence=0.95
    )

    response = LLMResponse(
        content="This is a generated answer.",
        citations=[citation],
        confidence=0.9,
        mode="chat",
        tokens_used=150,
        latency_ms=250.5,
    )
    print("LLMResponse and Citation instantiation successful!")
except Exception as e:
    print(f"Generation models test failed: {e}")
```

## 4. Test Exceptions

```python
try:
    raise IngestionError("Testing custom exception")
except IngestionError as e:
    print(f"Caught expected exception: {e}")
    assert isinstance(e, MiniNotebookLMError)
    print("Exception hierarchy verified!")
```
