# Phase 3: File Detection & Routing Test
This notebook verifies that the `FileDetector` can correctly identify various file types and URLs and route them to the appropriate handlers.

```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.ingestion.file_detector import FileDetector

print("Import successful!")
```

## 1. Test URL Detection (YouTube & Website)

```python
yt_url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"
web_url = "https://example.com"

yt_detect = FileDetector.detect(url=yt_url)
print(f"YouTube Detection: {yt_detect}")
assert yt_detect['source_type'] == 'youtube'
assert yt_detect['handler'] == 'youtube_pipeline'

web_detect = FileDetector.detect(url=web_url)
print(f"Website Detection: {web_detect}")
assert web_detect['source_type'] == 'website'
assert web_detect['handler'] == 'website_pipeline'
```

## 2. Test File Detection (Mocking files)

```python
# Create dummy files for testing
test_files = {
    "test.pdf": "pdf",
    "test.png": "image",
    "test.csv": "csv",
    "test.txt": "text"
}

for filename, expected_type in test_files.items():
    with open(filename, 'w') as f:
        f.write("dummy content")
    
    try:
        # Note: magic might not work perfectly with empty files, 
        # but the fallback to extension in FileDetector should handle it.
        detect = FileDetector.detect(file_path=filename)
        print(f"File {filename} detected as: {detect['source_type']}")
        assert detect['source_type'] == expected_type
    finally:
        if os.path.exists(filename):
            os.remove(filename)

print("File detection tests passed!")
```
