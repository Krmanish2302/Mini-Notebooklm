# Phase 6: Chunking System Test
This notebook verifies the different chunking strategies and the `ChunkerRegistry`.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.ingestion.chunking.chunker_registry import ChunkerRegistry
print("Import successful!")
```

## 1. Test Recursive Chunker (Default)

```python
text = "This is a long document. It has multiple sentences. We want to split it into chunks."
chunker = ChunkerRegistry.get_chunker("recursive", chunk_size=20, chunk_overlap=5)
chunks = chunker.chunk(text, {"source_id": "doc1"})

print(f"Recursive chunks: {len(chunks)}")
for c in chunks:
    print(f"  - {c['content']}")

assert len(chunks) > 1
```

## 2. Test Hierarchical Chunker

```python
text = "word " * 100
chunker = ChunkerRegistry.get_chunker("hierarchical", parent_size=50, child_size=10, overlap=2)
chunks = chunker.chunk(text, {"source_id": "doc2"})

print(f"Hierarchical chunks: {len(chunks)}")
print(f"First chunk metadata: {chunks[0]['metadata']}")
assert "parent_id" in chunks[0]['metadata']
```

## 3. Test Chunker Registry

```python
strategies = ChunkerRegistry.list_strategies()
print(f"Available strategies: {strategies}")
assert "recursive" in strategies
assert "semantic" in strategies
```
