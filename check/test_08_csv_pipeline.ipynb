## 1. Test CSV Processing
We'll create a dummy CSV and verify the extracted description.

```python
sample_csv = "data.csv"
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["NY", "SF", "LA"]
})
df.to_csv(sample_csv, index=False)

try:
    result = CSVPipeline.process(sample_csv, "source_csv_1")
    print("CSV Processing successful!")
    print(f"Description snippet: {result['content'][:200]}")
    
    assert "3 rows x 3 columns" in result['content']
    assert "Alice" in result['content']
    print("CSV test passed!")
except Exception as e:
    print(f"CSV test failed: {e}")
finally:
    if os.path.exists(sample_csv):
        os.remove(sample_csv)
```
