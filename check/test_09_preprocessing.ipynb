# Phase 5: Adaptive Preprocessing Test
This notebook verifies that the `AdaptivePreprocessor` correctly cleans content and recommends the optimal chunking strategy.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.ingestion.preprocessing.adaptive_preprocessor import AdaptivePreprocessor
print("Import successful!")
```

## 1. Test PDF Cleaning and Recommendation

```python
pdf_content = "[Page 1]\nThis is some text with hy-\nphenation. [Page 2]\nMore text here."
preprocessor = AdaptivePreprocessor()

result = preprocessor.process(pdf_content, "pdf")
print(f"Cleaned PDF Content: {repr(result['cleaned_content'])}")
print(f"Recommendation: {result['recommendation']}")

assert "hyphenation" in result['cleaned_content']
assert "## Page 1" in result['cleaned_content']
```

## 2. Test Website Cleaning

```python
web_content = "Main article content. All rights reserved. Subscribe to our newsletter."
result = preprocessor.process(web_content, "website")
print(f"Cleaned Website Content: {repr(result['cleaned_content'])}")
assert "All rights reserved" not in result['cleaned_content']
assert result['recommendation']['strategy'] == "topic_based"
```

## 3. Test YouTube Cleaning

```python
yt_content = "[00:15:30] Hello everyone. ♪ music playing ♪ [00:16:00] Next topic."
result = preprocessor.process(yt_content, "youtube")
print(f"Cleaned YouTube Content: {repr(result['cleaned_content'])}")
assert "music playing" not in result['cleaned_content']
assert result['recommendation']['strategy'] == "semantic"
```
